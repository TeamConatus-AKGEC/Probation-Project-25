```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("EDA.csv")

print("Shape of dataset:", df.shape)
print("\nColumns:", df.columns)
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
print("\nSummary statistics:\n", df.describe().T.to_markdown(index=False , numalign="left",stralign="left"))

print("\nFirst 5 rows:\n", df.head().to_markdown(index=False , numalign="left",stralign="left"))

print("\nUnique Crops:", df['Crop'].nunique())
print("Unique Seasons:", df['Season'].unique())
print("Unique States:", df['State'].nunique())

df[['Area','Production','Annual_Rainfall','Fertilizer','Pesticide','Yield']].hist(
    figsize=(12,8), bins=30, edgecolor='black'
)
plt.suptitle("Distributions of Numeric Features")
plt.show()

plt.figure(figsize=(10,5))
df['Crop'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Most Common Crops")
plt.xlabel("Crop")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
df.groupby("Season")['Yield'].mean().plot(kind='bar', color='skyblue')
plt.title("Average Yield by Season")
plt.ylabel("Yield")
plt.show()

plt.figure(figsize=(12,6))
df.groupby("State")['Production'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 States by Average Production")
plt.ylabel("Production")
plt.show()
```
